--> <font size="5"><b> Causal Inference </b> <br></font>
It is a process of estimating the causal effect of one variable on another variable from <b>observational data</b> <br>
<br>
<b>Causal effect</b> estimates aims at capturing strength of change (expected) in the Outcome variable when we modify the value of Treatment. <br>
In order to obtain an <b>unbiased estimate</b> of the causal effect, we need to get rid of <b>confounding</b>. 

--> we will learn <b>Heterogeneous Treatment Effect</b>, <br>
It indicates how treatment (T) effect vary with observable features of individual. It is used when the treatment assignment process is unknown.
 
Assume, data is generated from some collection policy. <br>
${Y_{i}, T_{i}, X_{i}, W_{i}, Z_{i}}$
    
$Y_{i}$ = Outcome <br> 
$T_{i}$ = Treatment/Decision <br>
$X_{i}$ = co-variates used for heterogeneity <br>
$W_{i}$ = Controls/Confounders/Common Cause ~ which effect on T and Y <br>
$Z_{i}$ = Instrumental Variable ~ which effect on T but not on Y <br>

<br>
** <b>Correlation does not imply causation.</b> <br>

A variable, X, can be said to cause another variable Y, if when all confounders are adjusted, <br>
an intervention in X results in a change in Y, but an intervention in Y does not necessarily result in a change in X. <br>
This is in contrast to correlations, which are inherently symmetric i.e. if X correlates with Y, Y correlates with X, while if X causes Y, Y may not cause X.<br>
Correlations are often driven by <b>confounding factors.</b> <br>
<br>
This leads to difference between <b> Bayesian </b> and <b> Causal Network </b> <br>
Case-1: <br>
Bayesian: P(Smoke|Fire) <br>
Causal: Fire causes Smoke <br>
Case-2: <br>
Bayesian: P(Fire|Smoke) <br>
Causal: Smoke causes Fire <br>

In Case-1, both are valid and can be estimated but in Case-2, bayesian is valid but causal is not <br>

Some <b>examples</b> why considering co-relation is bad idea: <br>
1) Basketball players are tall, but does playing basketball make you taller?

<b> --> <font size="3">Some Definitions: </b></font> <br>

<b>Group:</b> <br>
<b>Control group:</b> <br>
It receives either no treatment, a standard treatment whose effect is already known, or a placebo (a fake treatment to control for placebo effect). <br>
<b>Treatment group (experimental group):</b> <br>
It receives the treatment whose effect the researcher is interested in. <br>


<b>Outcome and Potential Outcomes:</b> <br>
$Y_{i}$ : it is the observed outcome of sample i. So this is factual (actually happened and present in observation). <br>
<br>
On the other hand, Potential Outcomes is related with treatment. <br>
$Y_{i, T=0}$: potential outcome when treatment is 0 <br>
$Y_{i, T=1}$: potential outcome when treatment is 1 <br>
<br>
So, if $Y_{i, T=0}$ is factual then $Y_{i, T=1}$ is counterfactual and vice-versa. <br>


<b>Causal Effect:</b> <br>
Causal effect of the treatment on the outcome for unit i is the difference between its two potential outcomes: <br>
$
\begin{align}
& \tau_{i} = Y_{i, T=1} - Y_{i, T=0}
\end{align}
$

<br>
<br>
<b>Average Treatment Effect(ATE):</b> <br>
It is calculated as the effect of the treatment on the outcome <br>
T --ate--> Y <br>
$
\begin{align}
& \tau_{ATE} = \frac{1}{N}\sum_{i}^{N}Y_{i, T=1} - \frac{1}{N}\sum_{i}^{N}Y_{i, T=0} \\
& = E[Y(T = 1) — Y(T = 0)] \\
& = E[\tau_{i}]
\end{align}
$

<br>
<br>
<b>Conditional Average Treatment Effect (CATE):</b> <br>
It is calculation of heterogeneity of treatment effects. It is the effect of the treatment on the outcome, conditional on confunders. <br>
T --ate--> Y <br>
Covariates(x) --cate--> Y <br>

$
\begin{align}
ATE = E[Y(T = 1)|X=x] — [Y(T = 0)|X=x]
\end{align}
$

<br>
<br>
<b>Marginal Effect:</b> <br>
The marginal effect of T (treatment) is just first derivative of outcome w.r.t to T. <br>
The rate at which Y changes given a one-unit increase in T holding all else constant.

--> <font size="3"><b> Difficulties of Gausal Inference: </b> </font><br>
A common problem with causal analysis throughout most sciences is that it requires experimentation which is not possible on observational data. <br>
This would require us to act one way and observe the consequences, and then go back in time and act another way to observe how the results may have been different. <br>
<br>
1) We can only observe particular change in T which causes an effect on Y - the <b>factual</b> <br>
2) We can never observe some other change in T which will cause different effect on Y - the <b>counterfactual</b> <br>

a) we can never directly calculate the counterfactual's effect, <br>
b) We can only estimate the counterfactuals. Causal Inference helps to estimate the <b>counterfactual.</b> <br>
c) expect some challenges when doing validation <br>

In other words, <br>
Where the action T was taken and Y observed [do(T=1)] ~ <b>factual</b> <br>
Where the action T was not taken (but everything else is the same) [do(T=0)] ~ <b>counterfactual</b>
<br>
<br>

<b>Observational Data and problems associated with it</b> <br>
In the context of Causal Inference data can be classified into two types: <br>
<br>
A) Randomized Control Trials (RCT) <br>
RCT is the traditional gold standard that enables <b>unbiased estimation</b> of treatment effects. <br>
1) In RCT, treatment is assigned to individuals randomly <br>
2) RCTs are often small datasets. <br>
3) They have limited generalizability that is there is a risk if participants are not representative of the population. <br>
4) RCT can be unethical in many cases such as Healthcare. <br>
5) They are unconfounded by design and thus we don’t have to deal with confounding. <br>

<br>

B) Observational Data <br>
1) Data is generated without the causal question in mind. <br>
2) Observational data is generally large and rich in nature. <br>
3) Observational data may represent the population. <br>
4) It may have a potential problem with hidden confounding. <br>

<br>
In the real world, observational data is the kind of data we work with and there are two challenges with observational data as follows: <br>

A) Finding Causal Direction <br>
From observational data, it is important to understand the causal direction. <br>
B) Confounding <br>
Confounder: A confounder is a variable that influences both treatment and target variables. Confounding can limit the identifiability of causal effect. <br>

Most current machine learning tools are designed to forecast what will happen next under the present strategy, but cannot be interpreted to <br>

For Causal Inference, Existing solutions to answer what-if questions are expensive. Decision-makers can engage in active experimentation like <br>

<b> <font size=3>--> Different Causal Estimation Methods </font> </b> 
<br>

1)Estimation methods under <b>Unconfoundedness</b> <br>
<b>Unconfoundedness</b> means all the potential variables (confounders) that could simultaneously have affected the treatment (T) and the outcome (Y) are observed. 
<br>
<br>
1.a) <b>Double Machine Learning (DML)</b> <br>
Doubly robust learning is a method for estimating heterogeneous treatment effects when, <br>
i)  the treatment is <b>both discrete and continuous  <br>
ii) all potential confounders/controls are observed  <br>

Double Machine Learning (DML) is an algorithm that applies arbitrary machine learning methods to fit the treatment and response, <br>
then uses a linear model to predict the response residuals from the treatment residuals. <br>
$
\begin{align}
& \text{model_y:} \\
& \bar{Y} = Y - E[Y|X, W] \\
& \bar{Y} \text{is residual} \\
& \text{model_t:} \\
& \bar{T} = T - E[T|X, W] \\
& \bar{Y} \text{is residual} \\
& \text{model_final:} \\
& \text{estimator for } \bar{Y} \text{ and } \bar{T}
\end{align}
$

1.b) <b>Doubly Robust Learning (DR)</b> <br>
Doubly robust learning is a method for estimating heterogeneous treatment effects when, <br>
i)  the treatment is <b>categorical  <br>
ii) all potential confounders/controls are observed  <br>
<br>
$
\begin{align}
& \text{model_regression: (Predicting the outcome from the treatment and controls)} \\
& \bar{Y} = Y - E[Y|T, X, W] \\
& \bar{Y} \text{is residual} \\
& \text{model_propensity: (Predicting the treatment from the controls)} \\
& \bar{T} = T - E[T|X, W] \\
& \bar{Y} \text{is residual} \\
& \text{model_final:} \\
& \text{estimator for } \bar{Y} \text{ and } \bar{T}
\end{align}
$

1.c)<b> Meta-Learners</b> <br>
The treatment space must be discrete. <br>

1.c.i)  <b>S-Learner</b> <br>
estimates the target variable using all the covariate features and treatment indicator, without giving the treatment indicator any special role. The S-Learner models Y(0) and Y(1) through one model. <br>
Stage-1 model: <br>
Estimate the average outcome $\mu(x)$ with covariates X and an indicator variable for treatment effect T: <br>
$
\begin{align}
& \mu(x) = E[Y|X=x, T=t] \\
\end{align}
$
<br>
Stage-2 model: <br>
Define the CATE estimate as: <br>
$
\begin{align}
& \hat{\tau}(x) = \hat{\mu}(x, T=1) - \hat{\mu}(x,T=0) \\
& = E[Y|X=x, T=1] - E[Y|X=x, T=0]
\end{align}
$

1.c.ii) <b>T-Learner</b> <br>
It estimates the response functions separately for the treatment and control populations. The T-Learner models Y(0) and Y(1) separately. <br>
Stage-1 model: <br>
Estimate the average outcome $\mu_{0}(x)$ and $\mu_{1}(x)$ <br>
$
\begin{align}
& \mu_{0}(x) = E[Y(0)|X=x] \\
& \mu_{1}(x) = E[Y(1)|X=x] \\
\end{align}
$
<br>
Stage-2 model: <br>
Define the CATE estimate as: <br>
$
\begin{align}
& \hat{\tau}(x) = \hat{\mu_{1}}(x) - \hat{\mu_{0}}(x) \\
& = E[Y(1)|X=x] - E[Y(0)|X=x] \\
& = E[Y(1) - Y(0)|X=x]
\end{align}
$

1.c.iii) <b>X-Learner</b> <br>
Stage-1 model: <br>
Estimate the average outcome $\mu_{0}(x)$ and $\mu_{1}(x)$ <br>
$
\begin{align}
& \mu_{0}(x) = E[Y(0)|X=x] \\
& \mu_{1}(x) = E[Y(1)|X=x] \\
\end{align}
$
<br>
Stage-2 model: <br>
Impute the user-level treatment effects  <br>
$D_{i}^{1}$ for user i in the treatment group based on $\mu_{0}(x)$ and <br>
$D_{j}^{0}$ for user j in the control groups based on $\mu_{1}(x)$  <br>
$
\begin{align}
& D_{i}^{1} = Y_{i}^{1} - \mu_{0}(x) \\
& D_{j}^{0} = \mu_{1}(x) - Y_{j}^{0} \\
\end{align}
$
<br>
Then estimate, <br>
$
\begin{align}
& \tau_{1}(x) = E[D_{1} | X=x] \\
& \tau_{0}(x) = E[D_{0} | X=x] \\
\end{align}
$
<br>
Stage-3 model: <br>
Define the CATE estimate by a weighted average of $\mu_{1}(x)$ and $\mu_{0}(x)$ <br>
$
\begin{align}
& \tau_{1}(x) = g(x)\tau_{0}(x) + (1-g(x))\tau_{1}(x) \\
\end{align}
$
<br>
where $g ∈ [0,1]$ is a weighting function.

Causality is represented mathematically via <b>Structural Causal Models (SCMs).</b> <br>
The two key elements of SCMs are a graph and a set of equations. <br>
More specifically, <br>
the graph is a <b>Directed Acyclic Graphs (DAG)</b> and the set of equations is a <b>Structural Equation Model (SEM).</b> <br>

<b> Four steps in Causal Inference </b> <br>

i/p: full causal graph and data <br>
1) Causal Discovery/Modeling: <br>
    o/p: find causal graph <br>
2) Causal Identification: <br>
    o/p: formulate what to estimate <br>
3) Causal Estimation: <br>
    o/p: compute estimation. causal quantities such as ATE, CATE. <br>
4) Causal Validation/Refutation: <br>
    o/p: robustness/validation scores for estimates <br>

(Recommendation Problem)
Question: 
    An online business would like to test a new feature or offering of their website and learn its effect on downstream revenue. Furthermore, <br>
    they would like to know which kind of users respond best to the new version.
Problem:    
    Ideally, the business would run A/B tests between the old and new versions of the website. However, a direct A/B test might not work because <br>
    the business cannot force the customers to take the new offering. Measuring the effect in this way will be misleading since not every customer <br>
    exposed to the new offering will take it. <br>
    The business also cannot look directly at existing data as it will be biased: the users who use the latest website features are most likely the <br>
    ones who are very engaged on the website and hence spend more on the company's products to begin with. Estimating the effect this way would be <br>
    overly optimistic. <br>
